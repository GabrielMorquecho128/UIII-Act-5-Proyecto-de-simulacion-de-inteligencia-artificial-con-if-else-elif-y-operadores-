```python
# ======================================================
# 🛡️ CBTis 128
# 📚 Elabora Proyectos de Programación Lógica
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Eliseo Nava
# 👨‍🎓 Alumno: Marco Morquecho
# 🛡️ Proyecto: IA para Guardias
# ======================================================

print("========= SISTEMA IA DE GUARDIAS 🛡️ =========")

nombre = input("Ingresa nombre del guardia: ")

zona = input("Ingresa zona de vigilancia: ")

horas = int(input("Ingresa horas trabajadas: "))

incidentes = int(input("Ingresa cantidad de incidentes: "))

acceso = input("¿Tiene acceso autorizado? si/no: ")

# ======================================================
# 🧠 Inteligencia Artificial
# ======================================================

if zona == "Banco":

    print("Zona de máxima seguridad")

if horas >= 12:

    print("Guardia con turno largo")

elif horas >= 8:

    print("Guardia con turno normal")

else:

    print("Guardia con turno corto")

if incidentes == 0 and acceso == "si":

    print("La IA detectó vigilancia segura")

elif incidentes >= 1 and incidentes <= 3:

    print("La IA detectó incidentes menores")

else:

    print("La IA detectó zona de riesgo")

# ======================================================
# 📊 Resultado final
# ======================================================

print("\n========= RESULTADOS =========")

print("Nombre:", nombre)

print("Zona:", zona)

print("Horas trabajadas:", horas)

print("Incidentes:", incidentes)

print("Acceso autorizado:", acceso)

print("\nPrograma finalizado 🛡️")
```
